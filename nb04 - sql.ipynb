### Using Parameterized Queries
A safer and more robust way to handle these issues is to use parameterized queries. This approach helps prevent SQL injection attacks and handles special characters automatically.

- Parameterized Query: The text function from SQLAlchemy is used to create a parameterized query. The :title_pattern placeholder is used for the parameter.
- Parameters: The params dictionary contains the actual value for the :title_pattern placeholder.
- Execution: The pd.read_sql function executes the query with the provided parameters, ensuring that the parameters are safely included in the query.

````python
import os
from sqlalchemy import create_engine, text # text is used to write parameterized queries
import pandas as pd
from dotenv import load_dotenv

# Load environment variables from .env file
load_dotenv()

# Connect to the database using SQLAlchemy's create_engine function
connection_string = f"postgresql://{os.getenv('DB_USER')}:{os.getenv('DB_PASSWORD')}@{os.getenv('DB_HOST')}/{os.getenv('DB_NAME')}"
engine = create_engine(connection_string)

# Define the SQL query using a parameterized query
query = text(""" -- Use a parameterized query to find all books with "Minds" in the title
SELECT
    title
FROM
    books
WHERE
    title LIKE :some_title
    AND author = :some_author
    AND date = :published_date
""")
````

````python

# Define the parameters
params = {"some_title": "%THE NAME OF THE BOOK%,
          "some_author": "THE NAME OF THE AUTHOR",
         }


# Use pandas to print one of the tables as a DataFrame using read_sql function
results = pd.read_sql(query, engine, params=params)
results
````
